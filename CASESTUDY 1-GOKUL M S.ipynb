# 3. Plot a bar graph of male vs female buyers participated in the sales.

```python
df['Buyer Gender'].value_counts()
```

```python
plt.figure(figsize=(8,8))
x=['MALE','FEMALE']
h=[4948,5052]
plt.bar(x,h,color='yellow')
plt.title('MALE vs FEMALE BUYERS PARTICIPATED IN THE SALES')
plt.xlabel('GENDER')
plt.ylabel('COUNT')
plt.grid(b = True, color ='black',
        linestyle ='-.', linewidth = 0.8,
        alpha = 0.7)
plt.show()
```
